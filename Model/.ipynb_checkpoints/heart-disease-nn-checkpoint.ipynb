```python
def draw_histograms(dataframe, features, rows, cols):
    fig=plt.figure(figsize=(20,20))
    for i, feature in enumerate(features):
        ax=fig.add_subplot(rows,cols,i+1)
        dataframe[feature].hist(bins=20,ax=ax,facecolor='midnightblue')
        ax.set_title(feature+" Distribution",color='DarkRed')
        
    fig.tight_layout()  
    plt.show()
draw_histograms(heart_df,heart_df.columns,6,3)
```

Some key inferences about the Framingham Heart Study dataset:

1. Gender Distribution:
   - About 44.5% of the participants are male (mean of Sex_male is 0.4452), indicating a fairly balanced gender distribution with slightly more females.

2. Age:
   - The average age is about 49.6 years, with a range from 32 to 70 years.
   - The middle 50% of participants are between 42 and 56 years old.

3. Smoking:
   - Approximately 48.8% of participants are current smokers.
   - Among smokers, the average number of cigarettes per day is about 9, but this varies widely (std dev of 11.9).

4. Blood Pressure Medication:
   - Only about 3% of participants are on blood pressure medication.

5. Prevalent Conditions:
   - Stroke is rare in the sample, with only about 0.56% having had a stroke.
   - Hypertension is more common, affecting about 31.2% of participants.
   - Diabetes is present in about 2.7% of the sample.

6. Cholesterol:
   - The average total cholesterol is 236.9 mg/dL, which is borderline high.
   - There's a wide range, from 113 to 696 mg/dL.

7. Blood Pressure:
   - Mean systolic BP is 132.4 mmHg, which is in the prehypertensive range.
   - Mean diastolic BP is 82.9 mmHg, also bordering on hypertension.

8. BMI:
   - The average BMI is 25.8, which is just in the overweight category.
   - BMI ranges from 15.5 (underweight) to 56.8 (severely obese).

9. Heart Rate:
   - Average resting heart rate is 75.7 bpm, which is within the normal range.

10. Glucose:
    - Mean glucose level is 81.9 mg/dL, which is normal.
    - There's a wide range, with some very high values (up to 394 mg/dL).

11. 10-Year CHD Risk:
    - About 15.2% of participants developed coronary heart disease within 10 years.

These statistics suggest a diverse sample with varying levels of cardiovascular risk factors. The presence of both healthy individuals and those with risk factors makes this dataset suitable for developing a heart disease prognosis model.

```python
import sklearn
new_features = heart_df[['age','Sex_male','cigsPerDay','totChol','sysBP','glucose','TenYearCHD']]
x = new_features.iloc[:,:-1]
y = new_features.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=5)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score
import numpy as np
from sklearn.utils import class_weight

# Assuming 'x' is your input features and 'y' is your target variable (binary labels)

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Standardize the input features (optional but often recommended)
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Define the model
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(x_train.shape[1],)))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))  # Output layer with 1 neuron for binary classification

class_weight = {0: 1, 1: 2,2: 3}

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(x_train_scaled, y_train, epochs=50, batch_size=32, validation_split=0.1,class_weight=class_weight)



```

```python
# Evaluate the model on the test set
y_pred_probs = model.predict(x_test_scaled)
y_pred = (y_pred_probs > 0.3).astype(int)
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy on Test Set: {accuracy}')

# Calculate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:')
print(conf_matrix)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Calculate precision, recall, and F1-score
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f'Precision: {precision}')
print(f'Recall: {recall}')

print(f'F1-Score: {f1}')
```

# Save Model

**To save the model from your code, you can use the following steps:**

- First, make sure you have a directory where you want to save the model. Let's say you want to save it in a folder called "saved_models" in your current working directory.
- Create the directory if it doesn't exist:

```python
import os

if not os.path.exists('saved_models'):
    os.makedirs('saved_models')
```

- Now, save the model using the save method:

```python
model_path = os.path.join('saved_models', 'heart_disease_model.keras')
model.save(model_path)
print(f"Model saved to {model_path}")
```

**This will save your model in the Keras format (.keras file).**
- To load and use this model later, you can do:

```python
from tensorflow.keras.models import load_model

loaded_model = load_model('saved_models/heart_disease_model.keras')
```

```python
import joblib

scaler_path = os.path.join('saved_models', 'scaler.joblib')
joblib.dump(scaler, scaler_path)
print(f"Scaler saved to {scaler_path}")
```
