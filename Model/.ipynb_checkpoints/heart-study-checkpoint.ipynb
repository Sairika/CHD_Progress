### Handling Missing Values:
- This code creates a new DataFrame 'updated_df' and fills in the missing values with the mean value of each respective column.

**Heatmap to finding strong corelation and weak corelation**

````python
corr_matrix=updated_df.corr()
plt.figure(1, figsize=(15,15))
sns.heatmap(corr_matrix, cmap='BrBG', center=0, annot=True, square=True)
````

1. Creating the correlation matrix:
```python
corr_matrix = updated_df.corr()
```
This line calculates the correlation coefficients between all pairs of variables in your updated dataset and stores them in a new DataFrame called `corr_matrix`.

2. Setting up the plot:
```python
plt.figure(1, figsize=(15,15))
```
This line creates a new figure (plot) with a size of 15x15 inches, which gives you a large, square plot that's easy to read.

3. Creating the heatmap:
```python
sns.heatmap(corr_matrix, cmap='BrBG', center=0, annot=True, square=True)
```
This line uses seaborn's heatmap function to create the correlation heatmap. Let's break down the parameters:

- `corr_matrix`: The data to be plotted (your correlation matrix)
- `cmap='BrBG'`: The color scheme. 'BrBG' stands for Brown-Green, which is a diverging colormap good for showing positive and negative correlations.
- `center=0`: This centers the colormap at 0, which is useful for correlation matrices as correlations range from -1 to 1.
- `annot=True`: This adds the numerical correlation values to each cell of the heatmap.
- `square=True`: This ensures each cell in the heatmap is square-shaped.

The resulting heatmap will show:
- Strong positive correlations as dark green
- Strong negative correlations as dark brown
- Weak or no correlations as light colors near white

The diagonal will always be 1 (dark green) as it represents each variable's correlation with itself.

This heatmap allows you to quickly identify:
1. Which variables are strongly correlated (positively or negatively) with the target variable (TenYearCHD)
2. Which variables are strongly correlated with each other, which could indicate multicollinearity in your data

````python
# boxplot
plt.figure(figsize = (15, 15))
plt.boxplot(updated_df)
plt.show()
````

Box plots, also known as box-and-whisker plots, serve several important purposes in data analysis and visualization:

1. **Display of distribution**: Box plots provide a concise summary of the distribution of a dataset, showing key statistical measures in a single graph.

2. **Identification of outliers**: They clearly show outliers, which are data points that fall significantly above or below the majority of the data.

3. **Comparison of datasets**: Box plots allow for easy comparison of distributions between different groups or categories.

4. **Visualization of central tendency**: They show the median (middle value) of the data, which is resistant to outliers.

5. **Display of variability**: Box plots illustrate the spread of the data through the interquartile range (IQR) and the overall range.

6. **Skewness indication**: The relative positions of the median, upper and lower quartiles can indicate whether the data is skewed.

7. **Compact representation**: They provide a lot of information in a relatively small space, making them efficient for comparing multiple datasets side by side.

8. **Robustness**: Box plots are less affected by extreme values than some other plots, making them useful for datasets with outliers.

9. **Quick insights**: They allow for quick, at-a-glance understanding of key characteristics of a dataset.

In the context of your heart disease prognosis project, box plots could be useful for comparing different variables across groups (e.g., comparing blood pressure between those who developed heart disease and those who didn't), or for identifying outliers in your medical data that might need further investigation.

# **Now, Logistic Regression Apply Algorithm**

> **Fixed the Target variable, Here the Target variable is TenYearCHD**

````python
x = updated_df.drop(columns=['TenYearCHD'])
y = updated_df['TenYearCHD']
````

````python
# split the dataset into training and testing 

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
````

> **Initialize and train the Logistic Regression model**

````python
model = LogisticRegression()
model.fit(x_train, y_train)

# Predict test set
y_pred = model.predict(x_test)
````

> **Calculate Evaluation matrix**

````python
accuracy_log = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
````

````python
print(f"Accuracy Score: {accuracy_log}")
print(f"Precision Score: {precision}")
print(f"Recall Score: {recall}")
print(f"F1 Score: {f1}")

# confusion matrix
print('\nConfusion matrix for Logistic:')
print(confusion_matrix(y_test,y_pred))
````

# **Now for K-Nearest Neighbors (KNN)**

> **Train the KNN model**

````python
knn_model = KNeighborsClassifier(n_neighbors=8)
knn_model.fit(x_train, y_train)

# Predict test set for KNN
y_pred_knn = knn_model.predict(x_test)
````

> **Calculate Accuracy Result for KNN**

````python
accuracy_knn = accuracy_score(y_test, y_pred_knn)
precision = precision_score(y_test, y_pred_knn)
recall = recall_score(y_test, y_pred_knn)
f1 = f1_score(y_test, y_pred_knn)

print(f"Accuracy Score: {accuracy_knn}")
print(f"Precision Score: {precision}")
print(f"Recall Score: {recall}")
print(f"F1 Score: {f1}")

# confusion matrix
print('\nConfusion matrix for KNN:')
print(confusion_matrix(y_test,y_pred_knn))
````

> Compare the accuracy using a graph

````python
models = ['Logistic Regression', 'KNN']
accuracies = [accuracy_log, accuracy_knn]

plt.figure(figsize=(8, 6))
plt.bar(models, accuracies, color=['green', 'blue'])
plt.ylim(0, 1)
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Accuracy of Logistic Regression vs KNN')
plt.show()
````
